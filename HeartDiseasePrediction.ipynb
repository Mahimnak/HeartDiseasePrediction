**Importing the dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data collection and processing**

```python
#loading csv data file to a pandas dataframe
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
heart_data.head()
```

```python
heart_data.shape
```

```python
heart_data.describe()
```

**Splitting the features and the target**

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
print(Y)
```

**Splitting the data into training and test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Model Training**

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(testing_data_accuracy)
```

**Building a predictive system**

```python
#Sample data with the target value of '1' so if model predicts 1 then the prediction is correct.
input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)

input_data_asnparray = np.asarray(input_data)
reshaped_data = input_data_asnparray.reshape(1,-1)

prediction = model.predict(reshaped_data)
print(prediction)
```
